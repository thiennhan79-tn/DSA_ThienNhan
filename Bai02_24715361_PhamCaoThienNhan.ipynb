## II. CẤU TRÚC DỮ LIỆU

Mỗi cuốn sách được lưu trữ dưới dạng Dictionary với cấu trúc:

```python
{
    "id": 1,
    "title": "Tên sách",
    "author": "Tác giả",
    "year": 2024,
    "status": "available"  # available hoặc borrowed
}
```

````python
books = [
    {"id": 1, "title": "Lập trình Python cơ bản", "author": "Nguyễn Văn A", "year": 2020, "status": "available"},
    {"id": 2, "title": "Khoa học dữ liệu với Python", "author": "Trần Thị B", "year": 2021, "status": "borrowed"},
    {"id": 3, "title": "Machine Learning cơ bản", "author": "Lê Văn C", "year": 2022, "status": "available"},
    {"id": 4, "title": "Python nâng cao", "author": "Nguyễn Văn A", "year": 2023, "status": "available"},
    {"id": 5, "title": "Big Data với Python", "author": "Phạm Thị D", "year": 2024, "status": "borrowed"}
]

print("Đã khởi tạo dữ liệu mẫu thành công!")
print(f"Số lượng sách: {len(books)}")
````

````python
def display_all_books():
    """
    Hiển thị danh sách tất cả các sách trong thư viện
    In ra: ID, Tên sách, Tác giả, Năm xuất bản, Trạng thái
    """
    # code của bạn
    print("Danh sách:")
    print("-----")
    for b in books:
        print(f"id: {b['id']}")
        print(f"title: {b['title']}")
        print(f"author: {b['author']}")
        print(f"year: {b['year']}")
        print(f"status: {b['status']}\n")
display_all_books()
````

````python
def add_new_book(title, author, year):
    """
    Thêm sách mới vào thư viện
    - Tự động tạo ID
    - lấy giá trị truyền vô: title, author, year
    - Mặc định status = "available"
    """
    # code của bạn
    if books:
        id = books[-1]["id"] + 1
    else:
        id = 1
    b = {
        "id":id,
        "title":title,
        "author":author,
        "year":year,
        "status":"available"
    }

    books.append(b)
    print("Thêm sách thành công")
add_new_book("Lập trình Python nâng cao", "Nguyễn Văn D", 2022)
````

````python
display_all_books()
````

````python
def borrow_book():
    """
    Mượn sách từ thư viện
    - Nhập ID sách cần mượn
    - Kiểm tra status = "available"
    - Cập nhật status = "borrowed"
    - Thông báo kết quả
    """
    print("\n--- MƯỢN SÁCH ---")
    
    # code của bạn
    id = int(input("Nhập id sách cần mượn:"))
    found = False
    for b in books:
        if b["id"] == id:
            found = True
            if b["status"] == "available":
                b["status"] = "borrowed"
                print(f"Bạn đã mượn sách {b['title']}")
                
            else:
                print("Sách đã được mượn")
            break
    if not found:
        print("Không có sách này trong thư viện")
    

borrow_book()
````

````python
def return_book():
    """
    Trả sách cho thư viện
    - Nhập ID sách cần trả
    - Kiểm tra status = "borrowed"
    - Cập nhật status = "available"
    - Thông báo kết quả
    """
    print("\n--- TRẢ SÁCH ---")
    
    # code của bạn
    id = int(input("Nhập id sách cần trả:"))
    found = False
    for b in books:
        if b["id"] == id:
            found = True
            if b["status"] == "borrowed":  
                b["status"] = "available"
                print("Trả sách thành công")
            else:
                print("Sách này chưa được mượn")
            break
    if not found:
        print("Không có loại sách này trong thư viện")

return_book()
````

````python
def search_by_author():
    """
    Tìm kiếm sách theo tác giả
    - Nhập tên tác giả
    - Duyệt list và hiển thị tất cả sách của tác giả đó
    - Sử dụng lower() để không phân biệt hoa thường
    """
    print("\n--- TÌM SÁCH THEO TÁC GIẢ ---")
    
    # code của bạn 
    author = input("Nhập tên tác giả cần tìm:").lower()
    found = False
    for b in books:
        if b["author"].lower() == author:
            print(f"ID: {b['id']}, Title: {b['title']}, Author: {b['author']}, Year: {b['year']}, Status: {b['status']}")
            found = True

    if not found:
        print("Không tìm thấy sách của tác giả này")

search_by_author()

````

````python
def filter_by_year():
    """
    Lọc sách theo khoảng năm xuất bản
    - Nhập năm bắt đầu và năm kết thúc
    - Hiển thị các sách trong khoảng năm đó
    """
    print("\n--- LỌC SÁCH THEO NĂM XUẤT BẢN ---")
    
    # code của bạn
    s = int(input("Nhập năm bắt đầu:"))
    e = int(input("Nhập năm kết thúc:"))

    if s > e:
        print("Năm bắt đầu không lớn hơn năm kết thúc")
        return
    found = False

    for b in books:
        if s <= b["year"] <= e:
            print(f"ID: {b['id']}, Title: {b['title']}, Author: {b['author']}, Year: {b['year']}, Status: {b['status']}")
            found = True
    if not found:
        print("Chưa có sách trong trong khoảng năm đó")

filter_by_year()
````

````python
def show_statistics():
    """
    Hiển thị thống kê thư viện
    - Tổng số sách
    - Số sách đang available
    - Số sách đang borrowed
    - Tính phần trăm
    """
    print("\n--- THỐNG KÊ THƯ VIỆN ---")
    
    # code của bạn 
    total = len(books)
    ava = 0
    bor = 0
    for b in books:
        if b["status"] == "available":
            ava += 1
        else:
            bor += 1
    print("Tổng số sách thư viện hiện tại:", total)
    print("Số sách đang available:", ava)
    print("Số sách đang borrowed:", bor)

    if total > 0:
        print(f"% available: {ava / total * 100:.2f}%")
        print(f"% borrowed: {bor / total * 100:.2f}%")
    else:
        print("Thư viện chưa có sách")

show_statistics()
````

````python
def main_menu():
    # code của bạn
    while True:
        print("\n====== MENU QUẢN LÝ THƯ VIỆN ======")
        print("1. Hiển thị danh sách sách")
        print("2. Thêm sách mới")
        print("3. Mượn sách")
        print("4. Trả sách")
        print("5. Tìm sách theo tác giả")
        print("6. Lọc sách theo năm xuất bản")
        print("7. Thống kê thư viện")
        print("0. Thoát")

        choose = input("Chọn chức năng: ")

        if choose == "1":
            display_all_books()
        elif choose == "2":
            add_new_book()
        elif choose == "3":
            borrow_book()
        elif choose == "4":
            return_book()
        elif choose == "5":
            search_by_author()
        elif choose == "6":
            filter_by_year()
        elif choose == "7":
            show_statistics()
        elif choose == "0":
            print("Thoát chương trình!")
            break
        else:
            print("Lựa chọn không hợp lệ, vui lòng chọn lại!")
````

````python
main_menu()
````
