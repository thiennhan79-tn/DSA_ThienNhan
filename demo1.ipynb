```python
#dictionary
products = [{
    "id":1,
    "name":"iphone",
    "price":100,
    "description":"iphone best"
},
{
    "id":2,
    "name":"ipad",
    "price":120,
    "description":'ipad best'
}]

print(products)
```

```python
#xuat danh sach
def XuatDanhSach():
    print("Danh sach san pham:")
    print("-----")
    for p in products:
        print(f"Ma san pham: {p['id']}")
        print(f"Ten san pham: {p['name']}")
        print(f"Gia san pham: {p['price']}")
        print(f"Mo ta san pham: {p['description']}")
        print("----\n")

XuatDanhSach()
```

```python
#them 1 sp
def them_sp():
    id = products[-1]["id"] + 1 if products else 1
    ten = input("Nhap ten:")
    gia = int(input("Nhap gia:"))
    mota = input("Nhap mo ta:")

    sp = {
        "id":id,
        "name":ten,
        "price":gia,
        "description":mota
    }

    products.append(sp)

them_sp()
```

```python
#Xoa sp
def Xoasp():
    find_id = int(input("Nhap id muon xoa:"))
    found = False
    for p in products:
        if p['id'] == find_id:
            products.remove(p)
            print("Da xoa thanh cong")
            found = True
            break
    if not found:
        print(f"Khong co id {find_id} de xoa")
Xoasp()
```

```python
#Tim kiem sp theo id
def tim_theo_id():
    ma = int(input("Nhap id can tim:"))
    found = False

    for p in products:
        if p['id'] == ma:
            print("Da tim thay san pham:")
            print(p)
            found = True
            break
    if not found:
        print(f"Khong tim thay san pham co id la {ma}")
tim_theo_id()
```

```python
#Tim kiem san pham theo ten
def tim_theo_ten():
    ten = input("Nhap ten san pham can tim: ").lower()

    found = False
    for sp in products:
        if ten in sp["name"].lower():
            print("Tim thay:", sp)
            found = True

    if not found:
        print(f"Khong tim thay san pham co ten {ten}")

tim_theo_ten()
```

```python
#Cap nhat 
def capnhat():
    id = int(input("Nhap id can cap nhat:"))
    found = False
    for p in products:
        if p['id'] == id:
            name = input("Nhap ten moi:")
            price = input("Nhap gia moi:")
            description = input("Nhap mo ta moi:")
            if name:
                p['name'] = name
            if price:
                p['price'] = price
            if description:
                p['description'] = description
            print("Cap nhat thanh cong")
            found = True
    if not found:
        print("Khong tim thay")

capnhat()
```
