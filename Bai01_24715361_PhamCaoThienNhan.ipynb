## **III. Data init**

````python
# Danh sách sinh viên ban đầu
students = [
    {
        "id": "SV001",
        "name": "Nguyễn Văn An",
        "age": 20,
        "scores": {"SAD": 8.5, "DSA": 7.0, "BigData": 9.0}
    },
    {
        "id": "SV002",
        "name": "Trần Thị Bình",
        "age": 21,
        "scores": {"SAD": 6.5, "DSA": 7.5, "BigData": 6.0}
    },
    {
        "id": "SV003",
        "name": "Lê Văn Cường",
        "age": 19,
        "scores": {"SAD": 9.0, "DSA": 8.5, "BigData": 9.5}
    },
    {
        "id": "SV004",
        "name": "Phạm Thị Dung",
        "age": 20,
        "scores": {"SAD": 5.0, "DSA": 4.5, "BigData": 6.0}
    },
    {
        "id": "SV005",
        "name": "Hoàng Văn Em",
        "age": 22,
        "scores": {"SAD": 7.0, "DSA": 7.0, "BigData": 7.5}
    }
]


print(f"Số lượng sinh viên: {len(students)}")
````

````python
def calculate_average(scores):
    """
    Tính điểm trung bình từ dictionary điểm
    
    Args:
        scores (dict): Dictionary chứa điểm các môn
    
    Returns:
        float: Điểm trung bình
    """
    # code của bạn
    if not scores:  
        return 0.0
    return sum(scores.values()) / len(scores)


print(f"Điểm trung bình: {calculate_average(students[0]['scores']):.2f}")
````

## **IV. CÁC CHỨC NĂNG**

### 1: Hiển thị danh sách sinh viên

````python
def display_all_students():
    """
    Hiển thị danh sách tất cả sinh viên
    In ra: ID, Tên, Tuổi, Điểm trung bình
    """
    # code của bạn
    print("Danh sách tất cả sinh viên")
    print("------")
    for s in students:
        print(f"ID: {s['id']}")
        print(f"Tên: {s['name']}")
        print(f"Tuổi: {s['age']}")
        print(f"Điểm Trung Bình: {calculate_average(s['scores']):.2f}")
        print("------\n")

# Test 
display_all_students()
````

### 2: Thêm sinh viên mới

````python
def add_new_student():
    """
    Thêm sinh viên mới
    - Tự động tạo ID theo format SV001, SV002...
    - Nhập: name, age
    - Nhập điểm 3 môn: SAD, DSA, BigData
    """
    print("\n--- THÊM SINH VIÊN MỚI ---")
    
    # code của 
    id = f"SV{len(students) + 1:03d}"
    name = input("Nhập tên: ")
    age = int(input("Nhập tuổi: "))
    scores = {}
    scores["SAD"] = float(input("Nhập điểm SAD: "))
    scores["DSA"] = float(input("Nhập điểm DSA: "))
    scores["BigData"] = float(input("Nhập điểm BigData: "))

    new_students = {
        "id": id,
        "name": name,
        "age": age,
        "scores": scores
    }
    students.append(new_students)

    
    
add_new_student()
````

### 3: Cập nhật điểm

````python
display_all_students()
````

````python
def update_scores():
    """
    Cập nhật điểm sinh viên
    - Nhập ID sinh viên
    - Hiển thị điểm hiện tại
    - Cho phép cập nhật từng môn (Enter để bỏ qua)
    - Cập nhật dictionary scores
    """
    print("\n--- CẬP NHẬT ĐIỂM ---")
    
    # code của bạn
    found = False
    id = input("Nhập id muốn cập nhật: ")
    for s in students:
        if s['id'] == id:
            name = input("Nhập tên mới: ")
            age = int(input("Nhập tuổi mới: "))
            scores = {}
            scores["SAD"] = float(input("Nhập điểm SAD mới: "))
            scores["DSA"] = float(input("Nhập điểm DSA mới: "))
            scores["BigData"] = float(input("Nhập điểm BigData mới: "))

            if name:
                s['name'] = name
            if age:
                s['age'] = age
            if scores:
                s['scores'] = scores
            
            print(f"Cập nhật thành công {id}")
            found = True
    if not found:
        print(f"Không tìm thấy mã {id}")

update_scores()
````

### 4: Xóa sinh viên theo ID

````python
def delete_student():
    """
    Xóa sinh viên theo ID 
    """
    print("\n--- XÓA SINH VIÊN ---")
    # code của bạn
    find_id = input("Nhập id muốn xóa: ")
    found = False
    for s in students:
        if s['id'] == find_id:
            students.remove(s)
            print("Đã xóa thành công ")
            found = True
            break
    if not found:
        print(f"Không có {id} để xóa")
   
delete_student()
````

### 5: Tìm sinh viên theo tên

````python
def search_by_name():
    """
    Tìm sinh viên theo tên
    - Không phân biệt hoa thường
    - Tìm kiếm gần đúng (chứa từ khóa)
    """
    print("\n--- TÌM SINH VIÊN THEO TÊN ---")
    find_ten = input("Nhập tên muốn tìm: ").lower()
    found = False
    for s in students:
        if find_ten in s['name'].lower():
            print("Sinh viên tìm được: ")
            print(s)
            found = True
    if not found:
        print(f"Không tìm thấy sinh viên có {find_ten}")
    
    
search_by_name()
````

### 6: Xếp hạng sinh viên

````python
def rank_students():
    """
    Xếp hạng sinh viên theo điểm trung bình
    - Tính điểm TB cho tất cả sinh viên
    - Sắp xếp theo điểm TB giảm dần
    - Sử dụng sorted() với key=lambda
    - Hiển thị bảng xếp hạng có số thứ tự
    """
    print("\n--- BẢNG XẾP HẠNG SINH VIÊN ---")
    
    # code của bạn
    ranked = sorted(
        students,
        key=lambda s: calculate_average(s["scores"]),
        reverse = True
    )

    print("STT | ID | Tên | Điểm TB")
    print("-" * 40)

    for i, s in enumerate(ranked, start=1):
        avg = calculate_average(s['scores'])
        print(f"{i} | {s['id']} | {s['name']} | {avg:.2f}")



rank_students()
````

### 7: Thống kê theo mức điểm

````python
def statistics_by_grade():
    """
    Thống kê sinh viên theo mức điểm
    - Giỏi: điểm TB >= 8.0
    - Khá: 6.5 <= điểm TB < 8.0
    - Trung bình: 5.0 <= điểm TB < 6.5
    - Yếu: điểm TB < 5.0
    - Đếm số sinh viên mỗi loại
    """
    print("\n--- THỐNG KÊ THEO MỨC ĐIỂM ---")
    
    # code của bạn
    gioi = 0
    kha = 0
    trung_binh = 0
    yeu = 0

    for s in students:
        avg = calculate_average(s["scores"])

        if avg >= 8.0:
            gioi += 1
        elif avg >= 6.5:
            kha += 1
        elif avg >= 5.0:
            trung_binh += 1
        else:
            yeu += 1

    print(f"Sinh viên Giỏi: {gioi}")
    print(f"Sinh viên Khá: {kha}")
    print(f"Sinh viên Trung bình: {trung_binh}")
    print(f"Sinh viên Yếu: {yeu}")


        


statistics_by_grade()
````

### 8: Tìm sinh viên điểm cao nhất/thấp nhất

````python
def find_top_bottom_students():
    """
    Tìm sinh viên có điểm cao nhất và thấp nhất
    - Sử dụng max()/min() với key
    - Hiển thị thông tin chi tiết
    """
    print("\n--- SINH VIÊN ĐIỂM CAO NHẤT VÀ THẤP NHẤT ---")

    # code của bank 
    top_student = max(
        students, key=lambda s: calculate_average(s["scores"])
    )

    bottom_student = min(
        students, key=lambda s: calculate_average(s["scores"])
    )

    print("\nSinh viên điểm cao nhất:")
    print(f"ID: {top_student['id']}")
    print(f"Tên: {top_student['name']}")
    print(f"Tuổi: {top_student['age']}")
    print(f"Điểm TB: {calculate_average(top_student['scores']):.2f}")

    print("\nSinh viên điểm thấp nhất:")
    print(f"ID: {bottom_student['id']}")
    print(f"Tên: {bottom_student['name']}")
    print(f"Tuổi: {bottom_student['age']}")
    print(f"Điểm TB: {calculate_average(bottom_student['scores']):.2f}")
   
find_top_bottom_students()
````

````python
def main_menu():
    while True:
        print("\n" + "="*50)
        print("     CHƯƠNG TRÌNH QUẢN LÝ SINH VIÊN VÀ ĐIỂM")
        print("="*50)
        print("1. Hiển thị danh sách sinh viên")
        print("2. Thêm sinh viên mới")
        print("3. Cập nhật điểm")
        print("4. Xóa sinh viên theo ID")
        print("5. Tìm sinh viên theo tên")
        print("6. Xếp hạng sinh viên")
        print("7. Thống kê theo mức điểm")
        print("8. Tìm sinh viên điểm cao nhất/thấp nhất")
        print("0. Thoát")
        print("="*50)
        
        choice = input("Chọn chức năng (0-8): ").strip()
        
        if choice == "1":
            display_all_students()
        elif choice == "2":
            add_new_student()
        elif choice == "3":
            update_scores()
        elif choice == "4":
            delete_student()
        elif choice == "5":
            search_by_name()
        elif choice == "6":
            rank_students()
        elif choice == "7":
            statistics_by_grade()
        elif choice == "8":
            find_top_bottom_students()
        elif choice == "0":
            print("Bye !")
            break
        else:
            print("\n✗ Lựa chọn không hợp lệ! Vui lòng chọn lại.")

main_menu()
````
